1. # Gender and Survival

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

train = pd.read_csv('train.csv') 
sns.barplot(data=train, x='Sex', y='Survived', ci=None)
plt.title('Survival Rate by Gender')
plt.show()
```

### Insight: Women had a significantly higher survival rate compared to men.


# 2. Passenger Class (Pclass) and Survival


```python
sns.barplot(data=train, x='Pclass', y='Survived', ci=None)
plt.title('Survival Rate by Passenger Class')
plt.show()
```


### Insight: Passengers in higher classes (Pclass = 1) had better survival chances.


# 3. Age and Survival

```python
sns.histplot(data=train, x='Age', hue='Survived', kde=True, bins=30)
plt.title('Age Distribution and Survival')
plt.show()
```

### Insight: Younger passengers had better survival rates compared to older ones.


# 4. Family Size (SibSp, Parch) and Survival


```python
train['FamilySize'] = train['SibSp'] + train['Parch'] + 1
sns.barplot(data=train, x='FamilySize', y='Survived', ci=None)
plt.title('Survival Rate by Family Size')
plt.show()
```

### Insight: Passengers with small family sizes (1-4 members) had better survival rates than those traveling alone or with large families.

# 5. Fare and Survival


```python
sns.boxplot(data=train, x='Survived', y='Fare')
plt.title('Fare Distribution by Survival')
plt.show()

```

### Insight: Higher ticket fares correlate with higher survival chances.


# 6. Embarkation Port and Survival (Embarked)


```python
sns.barplot(data=train, x='Embarked', y='Survived', ci=None)
plt.title('Survival Rate by Embarkation Port')
plt.show()
```

### Insight: Passengers embarking from specific ports (e.g., Cherbourg) had higher survival rates.


# 7. Combined Insights (Multivariate Analysis)


```python
sns.catplot(data=train, x='Pclass', hue='Sex', col='Survived', kind='count')
plt.suptitle('Pclass and Gender Combined Survival Analysis')
plt.show()
```
